# Anatomy of a Matplotlib Figure

## Parts of a Matplotlib Figure

````
# - pyplot is used only for few activities such as figure and creation
# Figure - Whole area chosen for plotting
# Axes - Area where data is plottd
# Axis - Number-line like objects, which define graph limits.
# Artist - Every element on the figure in an aritst.

# ---------------terms---------------------------------#
# Markers
# Figure
# Axes
# Axis
# Grid
# Title
# Major tick labels
# Minor tick labels
# Legend
````

## Figure

````
# Figure refer to whole area or page on which everything is drawn.
# It include Axes, Axis, and other Artist elements

fig = plt.figure()

# Executing the above code does not display any figure
# You need to explicitly tell pyplot to display it
````

## Viewing a Figure

````
fig = plt.figure()
plt.show()

# This output shows *figure object*
# You will be able to view a picture only when a figure contains at least on Axes element.
````

## Axes

````
# An Axes object is associated with only one Figure.
# A Figure can contain one or more number of Axes elements.
# A Axes can contain two Axis or three Axis for 2D , 3D plots respectively 
````

## Creating an Axes

````
# Syntax : add_subplot(nrows,ncols,index)
fig = plt.figure()
ax = fig.add_subplot(111)
# plt.add_subplot(111) is same as fig.sub_plot(1,1,1)
plt.show()
#  Above code will generate a figure with one axes , ax
````

## Adjusting Figure Size

````
fig = plt.figure(figsize= (8,6))
# the size is to be mention in tuple
# Syntax : plt.figure(figsize= (width, height))
ax = fig.add_subplot(111)
plt.show()
````

````
#  Line Plot Example
fig = plt.figure(figsize = (8,6))
ax = fig.add_subplot(111)
ax.set(title = 'Avg. Daily Temperature in Jan 2018',xlabel= 'Day',ylabel= 'Temperature(in deg',xlim = ([0,30]),ylim = ([25,35]))
days = [1,5,8,12,15,19,22,26,29]
tem = [29.3,30.1,30.4,31.5,32.3,32.6,31.8,32.4,32.7]
ax.plot(days,tem, label = 'Daily Temp')
ax.legend()
plt.show()
````

### Common Parameters of 'plot' Function

````
# color : Sets the color of the line
# linestyle : Sets the line style e.g., solid,dashed,etc.
# linewidth : Sets the  thickness of a line.
# marker : Chooses the marker for data points, e.g., circle, triangle,etc.
# markersize : Sets the size of the chosen marker.
# label : Names the line, which will comes in legend.
````

````
fig = plt.figure(figsize = (8,6))
ax = fig.add_subplot(111)
ax.set(title = 'Avg. Daily Temperature in Jan 2018',xlabel= 'Day',ylabel= 'Temperature(in deg',xlim = ([0,30]),ylim = ([25,35]))
days = [1,5,8,12,15,19,22,26,29]
tem = [29.3,30.1,30.4,31.5,32.3,32.6,31.8,32.4,32.7]
ax.plot(days,tem, color = 'green',
        linestyle = ':',
        linewidth = 2.5,
        label = 'Daily Temp',
        marker = 'o',
        markersize = 10)

# linestyle support values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'

ax.legend()
plt.show()
````

### Plotting multiple lines

## Scatter plot

### Little theory on scatter plot

````
# Scatter plot is similar to line plot
# Scatter plot is used for showing how one variable is related with anohter.
# Scatter plot consists of data  points. If the spread of data points is linear, then two variables are highly correlated.
# Scatte fucntion is used for drawing scatter plots.
````

### Plotting Scatter Plot

````
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.set(title='Avg. Daily Temperature of Jan 2018',
      xlabel='Day', ylabel='Temperature (in deg)',
      xlim=(0, 30), ylim=(25, 35))
days = [1, 5, 8, 12, 15, 19, 22, 26, 29]
temp = [29.3, 30.1, 30.4, 31.5, 32.3, 32.6, 31.8, 32.4, 32.7]
ax.scatter(days,temp)
plt.show()
````

### Common Parameters of 'Scatte' plot

````
# c : Sets color of markers.
# s : Sets size of markers.
# marker : Select a marker. e.g., circle,triangle,etc.
#edgecolor : Sets the color o lines on edges of markers.
````

````
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.set(title='Avg. Daily Temperature of Jan 2018',
      xlabel='Day', ylabel='Temperature (in deg)',
      xlim=(0, 30), ylim=(25, 35))
days = [1, 5, 8, 12, 15, 19, 22, 26, 29]
temp = [29.3, 30.1, 30.4, 31.5, 32.3, 32.6, 31.8, 32.4, 32.7]
ax.scatter(days,temp,c = ['black'],s = 40,edgecolors=['green'],marker = '*')
plt.show()
# parameters 'c' and 's' takes list of values.
# If the no.of values is less than given values the list is repeated
````

# Bar Plot and Pie Plot

## Bar Plot

````
# Bar Plot is commonly used for comparing categories.
# It is also used to compare categories over a short period of time.
# bar and barh are used for plotting vertical and horizontal bar plots repectively

# Syntax : bar(x,height) 
# x - coordinates of bars.
# height - List of height of each bar

# Syntax - barh(y,width)
# y = y coordintes of bars
# widht - List of widths
````

### Bar Plot Using 'bar'

````
fig = plt.figure(figsize = (8,6))
ax = fig.add_subplot(111)
ax.set(title = 'Avg. Quarterly Sales',xlabel = 'Quater', ylabel = 'Sales (in millions)')
quarters = [1,2,3]
sales_2017 = [25782,35783,36133]
ax.bar(quarters,sales_2017)
ax.set_xticks(quarters)
ax.set_xticklabels(['Q1-2017','Q2-2017','Q3-2017'],rotation = 10)
# ax.set_yticklabels(range(0,35000,10000),rotation = 25)
plt.show()
````

### Common Parameters of 'bar'

````
# color : Sets the color of bars.
# edgecolor : Sets the color of the border line of bars
# width : Sets the width of bars
# align : Aligns the bars w.r.t. x-coordinates
# label : Sets label to a bar, appearing in legend
# linewidth : Sets the width of the line
````

````
fig = plt.figure(figsize = (8,6))
ax = fig.add_subplot(111)
ax.set(title = 'Avg. Quarterly Sales',xlabel= 'Quater',ylabel= 'Sales (in millions)')
s = [1,2,3]
sales_2017 = [25782,35783,36133]
ax.set_xticks(s)
ax.set_xticklabels(['Q1-2017','Q2-2017','Q3-2017'])
ax.bar(s,sales_2017,color ='salmon',edgecolor = 'black',width= 0.6,label = 'Sales',align = 'center',linewidth = 2,linestyle = '-.')
plt.show()

# the align key words are 'center', 'edge'
````

## Pie Plot

````
# Pie plot : is effective in showing the proportion of categories
# It is best suited used for comparing fewer categories
# In general Pie Plot is used highlight proportion of one or more categories
# Syntax pie(x)
# x = sizes of protion ,passed either as a fraction or  a number.
````

````
fig = plt.figure(figsize = (8,6))
ax = fig.add_subplot()
ax.set(title = 'Avg. Quarterly Sales')
sales_2017 = [25782,35783,36133]
ax.pie(sales_2017,shadow=True)
plt.show()
````

### Common Parameters of 'pie'

## Example of Using GridSpec

````
# The below example recreates the previous figure using GridSpec
# A GridSpec object, gd is created with two rows and two columns
# Then a selcted grid portion is passed as an argument to subplot

import matplotlib.gridspec as gridspec
fig = plt.figure(figsize = ( 10,8))
gd = gridspec.GridSpec(2,2)

ax1 = plt.subplot(gd[0,:],title = 'plot 1')
ax1.set(xticks = [],yticks = [])
ax2 = plt.subplot(gd[1,0],title = 'plot 2')
ax2.set(xticks  = [], yticks = [])
ax3 = plt.subplot(gd[-1,-1],title = 'plot 3')
ax3.set(xticks = [],yticks = [])
plt.show()
````

## Common Pitfalls in Data Visualization

````
# Common pitfalls to be avoided for a better Data Visualization are:
#   Creating unlabelled plots.

#   Using 3-Dimensional charts. Don't prefer 3-D plots, unless they add any value over 2-D charts.

#   Portions of a pie plot do not sum up to meaningful number.

#   Showing too many portions in a single pie chart.

#   Bar charts not starting at zero.

#   Failing to normalize the data.

#   Adding extra labels, and fancy images.
````

## Best Pracitces of Data Visualization

````
# Few of the best practices of Data Visualization are:

#   Display the data points on the plot, whenever required.

#   Whenever correlation is plotted, clarify that you have not established any cause of link between the variables.

#   Prefer labeling data objects directly inside the plot, rather than using legends.

#   Create a visualization, which stands by itself. Avoid adding extra text to tell more about visualization.
````
